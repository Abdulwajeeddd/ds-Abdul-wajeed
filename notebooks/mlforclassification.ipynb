```python
drop_cols = [
    'Account',
    'classification',       # target
    'Transaction Hash',
    'Order ID',
    'Trade ID',
    'Timestamp IST',
    'trade_time_utc',
    'Closed PnL'  # likely leakage
]
```

```python
X = df_class_train.drop(columns=drop_cols)
categorical_cols = X.select_dtypes('object').columns.tolist()
numeric_df = X.drop(columns=categorical_cols)
y = df['classification']
```

```python
numeric_cols
```

```python
categorical_cols
```

```python
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
encoded_array = encoder.fit_transform(X[categorical_cols])

encoded_df = pd.DataFrame(
    encoded_array,
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X.index
)

df_final = pd.concat([numeric_df, encoded_df], axis=1)
```

```python
print(df_final.dtypes)
```

```python
#splitting the data before fitting
X_train, X_test, y_train, y_test = train_test_split(df_final, y, test_size=0.2, random_state=42)
model = RandomForestClassifier(
    n_estimators=200,
    random_state=42,
    n_jobs=-1
)
model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
```

```python
# Select rows where classification is missing
X_pred = df_class_pred.drop(columns=drop_cols)
categorical_cols = X_pred.select_dtypes('object').columns.tolist()
numeric_df_pred = X_pred.drop(columns=categorical_cols)

# Encode categorical columns
encoded_array_pred = encoder.transform(X_pred[categorical_cols])
encoded_df_pred = pd.DataFrame(
    encoded_array_pred,
    columns=encoder.get_feature_names_out(categorical_cols),
    index=X_pred.index
)

df_final_pred = pd.concat([numeric_df_pred, encoded_df_pred], axis=1)
```

```python
# Predict classification for missing rows
predicted_class = model.predict(df_final_pred)
df.loc[df['classification'].isna(), 'classification'] = predicted_class
df.to_csv("filled_classification.csv", index=False)
```
